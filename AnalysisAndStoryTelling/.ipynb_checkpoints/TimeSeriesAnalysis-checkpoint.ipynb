# Time series analysis.

```python
import os
import pickle
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import datetime 

%matplotlib inline
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

pd.set_option('display.max_columns', None) 

# Setup Seaborn
sns.set_style("whitegrid")
sns.set_context("poster")

sns.set(rc={'figure.figsize':(15,10)})
```

```python
with open('../Data/Crisis_data_cleaned.txt', 'rb') as f:
    results = pickle.load(f)
```

```python
df = pd.DataFrame.from_records(results)
#df.info()
```

```python
occured_date_time = df[['occured_date_time']]
reported_date_time = df[['reported_date_time']]
```

```python
reported_date_time['15min_round_time'] = reported_date_time['reported_date_time'].dt.round('15min')
reported_date_time['hour_time'] = reported_date_time['reported_date_time'].dt.round('H')
reported_date_time['day_time'] = reported_date_time['reported_date_time'].dt.round('D')
reported_date_time['minute'] = reported_date_time['reported_date_time'].dt.minute
reported_date_time['hour'] = reported_date_time['reported_date_time'].dt.hour
reported_date_time['day'] = reported_date_time['reported_date_time'].dt.day
reported_date_time['month'] = reported_date_time['reported_date_time'].dt.month
reported_date_time['year'] = reported_date_time['reported_date_time'].dt.year
reported_date_time['weekday_name'] = reported_date_time['reported_date_time'].dt.weekday_name
reported_date_time.head()
```

```python
occured_date_time['15min_round_time'] = occured_date_time['occured_date_time'].dt.round('15min')
occured_date_time['hour_time'] = occured_date_time['occured_date_time'].dt.round('H')
occured_date_time['day_time'] = occured_date_time['occured_date_time'].dt.round('D')
occured_date_time['minute'] = occured_date_time['occured_date_time'].dt.minute
occured_date_time['hour'] = occured_date_time['occured_date_time'].dt.hour
occured_date_time['day'] = occured_date_time['occured_date_time'].dt.day
occured_date_time['month'] = occured_date_time['occured_date_time'].dt.month
occured_date_time['year'] = occured_date_time['occured_date_time'].dt.year
occured_date_time['weekday_name'] = occured_date_time['occured_date_time'].dt.weekday_name
occured_date_time.head()
```

```python
WINDOW=1000
def plot_series(ts, ts2, window, title, show_std=True):
    plt.figure(figsize=(20,8))
    plt.plot(pd.rolling_mean(ts,  window=window), label='Reported Mean')
    if show_std:
        plt.plot(pd.rolling_std(ts,  window=window), label='Reported Standard deviation')
    plt.plot(pd.rolling_mean(ts2,  window=window), label='Occurred Mean')
    if show_std:
        plt.plot(pd.rolling_std(ts2,  window=window), label='Occurred Standard deviation')
    plt.legend(loc='best')
    plt.title(title+' Rolling Mean & Standard Deviation.')
    plt.show(block=False)
```

### Plot time series for reported and occurred date times.

```python
reported_hour_time = reported_date_time.groupby('reported_date_time').size().reset_index().set_index('reported_date_time')[0]
occured_hour_time = occured_date_time.groupby('occured_date_time').size().reset_index().set_index('occured_date_time')[0]
plot_series(reported_hour_time, occured_hour_time, 672, 'Day time', False)
```

#### Plotting reported and occurred times, major spike happened around 2017-01 (i.e. more occurrences happened).

```python
reported_15min = reported_date_time.groupby('15min_round_time').size().reset_index().set_index('15min_round_time')[0]
occured_15min = occured_date_time.groupby('15min_round_time').size().reset_index().set_index('15min_round_time')[0]
plot_series(reported_15min, occured_15min, 672, '15 min rounding', False)
```

```python
reported_hour_time = reported_date_time.groupby('hour_time').size().reset_index().set_index('hour_time')[0]
occured_hour_time = occured_date_time.groupby('hour_time').size().reset_index().set_index('hour_time')[0]
plot_series(reported_hour_time, occured_hour_time, 168, 'Hour time', False)
```

```python
reported_days = reported_date_time.groupby('day_time')['day'].size().reset_index().set_index('day_time').day
occured_days = occured_date_time.groupby('day_time')['day'].size().reset_index().set_index('day_time').day
plot_series(reported_days, occured_days, 7, 'Day time', False)
```

#### From the above plots grouping by 15 min plots, hour time and day time we get to see the spikes for crisis occurred times (used moving/rolling mean, standard deviation to plot them).

#### There are many crisis occurrences happened during certain time periods.

```python
reported_month = reported_date_time.groupby('month').size().reset_index().set_index('month')[0]
occured_month = occured_date_time.groupby('month').size().reset_index().set_index('month')[0]
plot_series(reported_month, occured_month, 1, 'Month', False)
```

```python
reported_day = reported_date_time.groupby('day').size().reset_index().set_index('day')[0]
occured_day = occured_date_time.groupby('day').size().reset_index().set_index('day')[0]
plot_series(reported_day, occured_day, 1, 'Day', False)
```

#### Early month of the year and day has more occurrences.

```python
reported_hour = reported_date_time.groupby('hour').size().reset_index().set_index('hour')[0]
occured_hour = occured_date_time.groupby('hour').size().reset_index().set_index('hour')[0]
plot_series(reported_hour, occured_hour, 1, 'Hour', False)
```

```python
reported_hour = reported_date_time.groupby('weekday_name').size().reset_index().set_index('weekday_name')[0]
occured_hour = occured_date_time.groupby('weekday_name').size().reset_index().set_index('weekday_name')[0]
plot_series(reported_hour, occured_hour, 1, 'Weekday', False)
```
